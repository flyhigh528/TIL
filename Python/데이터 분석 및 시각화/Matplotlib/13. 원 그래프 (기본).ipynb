# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Max
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python','Java','JavaScript','C#','C/C++','ETC']
colors = ['r','g','b','magenta','yellow','orange']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python','Java','JavaScript','C#','C/C++','ETC']
# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.05] * 6

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title = '언어')
plt.title('언어 별 선호도')
plt.show()
```
